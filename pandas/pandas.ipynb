Manipulating a DataFrame

```python
#adding a column
wine_df['Hue'] = wine_dataset.target
```

```python
#removing a row/column
wine_df.drop(index=0,axis=1)    #index=target row/column-1 and axis=0 -> row axis=1 -> column
```

```python
#drop a column
wine_df.drop(columns ='Hue',axis=1)
```

```python
#locating a row using index value
wine_df.iloc[2]
```

```python
#locating a particular column
print(wine_df.iloc[:,0]) #first column
print(wine_df.iloc[:,1]) #second column
print(wine_df.iloc[:,2]) #third column
print(wine_df.iloc[:,-1]) #last column
```
