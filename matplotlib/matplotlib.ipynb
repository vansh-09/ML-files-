```python
#importing the lib
import matplotlib.pyplot as plt
```

```python
#import numpy to get value for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print(y)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x axis and y axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['English','French','Spanish']
people = [10,20,23]
ax.bar(lang,people)
plt.xlabel('Languages')
plt.ylabel('People')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
lang = ['English','French','Spanish']
people = [10,20,23]
ax.pie(people, labels=lang ,autopct="%1.1f%%")
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)

fig2=plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
